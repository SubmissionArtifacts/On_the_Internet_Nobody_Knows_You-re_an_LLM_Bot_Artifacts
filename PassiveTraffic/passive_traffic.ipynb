```python
import pandas as pd 
```

```python
df_passive = pd.read_csv("passive_data.csv")
print(df_passive.shape)
```

```python
# group by user-agent column and compute the total number of _id for each user-agent in a dataframe called bots_df
bots_df = df_passive.groupby("user-agent")["_id"].count().reset_index()
bots_df.rename(columns={"_id": "count"}, inplace=True)
# sort the result by the count of _id in descending order
bots_df = bots_df.sort_values(by="count", ascending=False).reset_index()
# display full col width of the dataframe
pd.set_option("display.max_colwidth", None)
bots_df.head(50)
```

```python
# select rows from df_passive where user-agent contains bot insensitve 
df_passive_bots = df_passive[df_passive["user-agent"].str.contains("bot", case=False, na=False)]
print(df_passive_bots.shape)
bots_df = df_passive_bots.groupby("user-agent")["_id"].count().reset_index()
bots_df.rename(columns={"_id": "count"}, inplace=True)
print(bots_df.shape)
# sort the result by the count of _id in descending order
bots_df = bots_df.sort_values(by="count", ascending=False).reset_index()
bots_df.to_csv("bots_user_agents.csv", index=False)
# display full col width of the dataframe
pd.set_option("display.max_colwidth", None)
bots_df.head(50)
```

```python
# group by siteKey column and compute the total number of _id for each siteKey in a dataframe called bots_df
bots_df = df_passive.groupby("siteKey")["_id"].count().reset_index()
bots_df.rename(columns={"_id": "count"}, inplace=True)
# sort the result by the count of _id in descending order
bots_df = bots_df.sort_values(by="count", ascending=False).reset_index()
# display full col width of the dataframe
pd.set_option("display.max_colwidth", None)
bots_df.head(50)
```
